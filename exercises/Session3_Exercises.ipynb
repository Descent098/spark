```python
%%ignite
# Pythonic Lava #3 - Drag and launch!

is_dragging = False

x = 0
y = 0
x_speed = 0
y_speed = 0

def setup():
    size(500, 500)
    
def draw():
    global is_dragging
    global x, y, x_speed, y_speed
    
    background("green")
    
    if mouse_is_pressed:
        if not is_dragging:
            is_dragging = True
            x = mouse_x
            y = mouse_y
            x_speed = 0
            y_speed = 0
        draw_line(x, y, mouse_x, mouse_y)
    else:        
        if is_dragging:
            scale = 0.1
            x_speed = scale * (x - mouse_x)
            y_speed = scale * (y - mouse_y)
            
        is_dragging = False
    
    x += x_speed
    y += y_speed
        
    fill_style("pink")
    circle(x, y, 30)
```

```python
%%ignite
# Pythonic Lava #4 - Pong!

# Importing random is not needed in a future version (tentative v0.0.14)
# If students opt to hard-code anything that uses random, then full points should be rewarded
# The ball changing colours randomly can also be omitted for full marks. It's just to make it a little more fun :P
import random

x = 250
y = 100
x_speed = 5
y_speed = 5
radius = 15

paddle_width = 120
paddle_height = 20
paddle_x = 250 - paddle_width / 2
paddle_y = 400

# Hue in hue, saturation, light. RGB works too, I'm just using HSL because you can get brighter colours
hue = 0
has_bounced = False  # Used to detect bounces and trigger colour change. Not needed for marks

def setup():
    size(500, 500)
    
def draw():
    global x, y, x_speed, y_speed, radius
    global paddle_width, paddle_height, paddle_x, paddle_y
    global hue, has_bounced
    background(0)
    
    x += x_speed
    y += y_speed
    
    paddle_x = mouse_x - paddle_width / 2
    
    has_bounced = False
    
    # Wall bounces
    if y - radius < 0:
        y_speed *= -1
        has_bounced = True
    if x - radius < 0 or x + radius > width:
        x_speed *= -1
        has_bounced = True
        
    # Paddle bounce
    if (x > paddle_x and x < paddle_x + paddle_width and
        y + radius > paddle_y and y - radius < paddle_y + paddle_height):
        y_speed *= -1
        has_bounced = True
    
    # Change colour - not needed for full marks
    if has_bounced:
        hue = random.randint(0, 360)
    
    fill_style(255)
    rect(paddle_x, paddle_y, paddle_width, paddle_height)
    
    fill_style("hsl({}, 100%, 64%)".format(hue))
    circle(x, y, radius * 2)
```
