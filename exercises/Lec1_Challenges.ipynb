```python
%%ignite
# Challenge 6
from math import sin

def setup():
    size(500, 500)
    
def draw():
    fill_style(120, 220, 255)
    fill_rect(0, 0, width, height)
    
    mid_x = width / 2
    mid_y = height / 2
    
    mouse_frequency = 2 * pi / 200
    offset_frequency = 2 * pi / 400
    x1 = 100
    x2 = 175
    x3 = 250
    x4 = 325
    h1 = 40 * sin(mouse_y * mouse_frequency + x1 * offset_frequency)
    h2 = 60 * sin(mouse_y * mouse_frequency + x2 * offset_frequency)
    h3 = 80 * sin(mouse_y * mouse_frequency + x3 * offset_frequency)
    h4 = 100 * sin(mouse_y * mouse_frequency + x4 * offset_frequency)
    
    fill_style(0, 0, 255)
    rect(x1, mid_y, 30, h1)
    
    fill_style(100, 100, 255)
    rect(x2, mid_y, 30, h2)
    
    fill_style(130, 170, 255)
    rect(x3, mid_y, 30, h3)
    
    fill_style(220, 220, 255)
    rect(x4, mid_y, 30, h4)
    
    
```
