```python
%%ignite

from math import sin, cos, pi

timer = 0

def setup():
    size(500, 500)
#     fill_style('red')
#     rect(100, 100, 20, 40)
    
def draw():
    global timer
    
    background('black')
    
    fill_style('yellow')
    circle(mouse_x, mouse_y, 60)
    
    
    freq0 = pi / 60
    x0 = mouse_x + 66 * sin(timer * freq0)
    y0 = mouse_y + 66 * cos(timer * freq0)
    
    fill_style(100, 255, 100)
    circle(x0, y0, 30)
    
    freq1 = pi / 80
    x1 = mouse_x + 133 * sin(timer * freq1)
    y1 = mouse_y + 133 * cos(timer * freq1)
    
    fill_style(150, 255, 150)
    circle(x1, y1, 30)
    
    freq2 = pi / 100
    x2 = mouse_x + 200 * sin(timer * freq2)
    y2 = mouse_y + 200 * cos(timer * freq2)
    
    fill_style(200, 255, 200)
    circle(x2, y2, 30)
    
    timer += 1
```

```python
%%ignite
from math import sqrt

def setup():
    global x1, y1, x1_speed, y1_speed
    global x2, y2, x2_speed, y2_speed
    size(500, 500)
    
    x1 = width / 3
    y1 = height / 2
    x1_speed = 0
    y1_speed = 0
    
    x2 = width  * 2 / 3
    y2 = height / 2
    x2_speed = 0
    y2_speed = 0
    
def draw():
    global x1, y1, x1_speed, y1_speed
    global x2, y2, x2_speed, y2_speed
    background(0, 0, 128)
    
    scale1 = 0.03
    x1_speed = scale1 * (mouse_x - x1)
    y1_speed = scale1 * (mouse_y - y1)
    x1 += x1_speed
    y1 += y1_speed
    
    scale2 = -1
    x2_dist = mouse_x - x2
    y2_dist = mouse_y - y2
    dist2 = sqrt( x2_dist ** 2 + y2_dist ** 2)
    
    x2_speed = scale2 * dist2 / x2_dist
    y2_speed = scale2 * dist2 / y2_dist
    x2 += x2_speed
    y2 += y2_speed
    
    fill_style(100, 255, 100)
    circle(x1, y1, 40)
    
    fill_style(255, 0, 0)
    circle(x2, y2, 40)
    
```

```python
%%ignite

from math import hypot

x1 = 0
x1_speed = 8
y1 = 250
y1_speed = 0

def setup():
    size(500, 500)
    
def draw():
    global x1, y1, x1_speed, y1_speed
    
    background(160, 255, 160)
    
    dist = hypot(mouse_x - x1, mouse_y - y1)
    red = int(dist / width * 255 * 1.5)
    
    stroke_style(red, 0 ,0)
    line_width(3)
    draw_line(mouse_x, mouse_y, x1, y1)
    
    scale1 = 0.001
    x1_speed += scale1 * (mouse_x - x1)
    x1 += x1_speed
    y1_speed += scale1 * (mouse_y - y1) 
    y1 += y1_speed
    
    fill_style(100, 100, 255)
    stroke_style('black')
    canvas.line_width = 1
    circle(x1, y1, 20)
```

```python
%%ignite

x = 30
y = 50
x_speed = 3
y_speed = -3
gravity = 0.7

def setup():
    size(500, 500)
    
def draw():
    global x, y, x_speed, y_speed, gravity
    background(100, 100, 255)

    x += x_speed
    x %= width
    y += y_speed
    y_speed += gravity
    
    magnet_y = 400
    magnet_force = -100
    dist_y = magnet_y - y
    y_force = magnet_force / (dist_y)
    y_speed += y_force
    
    y_color = int(y_force * -128)
    
    fill_style(y_color + 80, 128 - y_color, 128 - y_color)
    rect(0, magnet_y, width, 40)
    fill_style(255)
    rect(100, magnet_y + 18, 20, 4)  
    rect(width / 2 - 10, magnet_y + 18, 20, 4)  
    rect(380, magnet_y + 18, 20, 4)  
    
    fill_style('red')
    circle(x, y, 40)
    
    fill_style('white')
    rect(x - 10, y - 2, 20, 4)    
```
